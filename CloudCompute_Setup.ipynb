# AWS Console Setup Guide for the Cloud Research Workshop

This notebook is a **guide for the AWS Console setup phase** before the command-line workflow starts.

Use this notebook while working in the AWS web interface. It is intentionally light on code and focuses on what to click, what to choose, and why.

## Goal
By the end of this setup, you should have:
- one running EC2 instance,
- one SSH key pair downloaded on your laptop,
- a security group that allows SSH from your own IP,
- the instance ID and public IP copied into your CLI notebook,
- an informed decision about whether your root volume should be deleted on termination.

## Step 1 — Log in and select the region

1. Open the AWS Management Console.
2. In the top search bar, type **EC2** and open the EC2 service.
3. In the upper-right corner, choose the workshop region, for example **eu-central-1 (Frankfurt)**.

### Why this matters
- Regions are physically different AWS locations.
- Your instance, storage, and networking resources live in the selected region.
- If you accidentally switch regions later, your instance may seem to “disappear” even though it still exists elsewhere.

### Good practice
Write your selected region down now. You will need the same region later in the CLI notebook.

```python
REGION = "eu-central-1"
print(f"Selected workshop region: {REGION}")
```

## Step 2 — Open the EC2 launch wizard

1. In the EC2 dashboard, click **Launch instance**.
2. Keep the launch wizard open and work through the sections from top to bottom.

### Suggested naming convention
Use a clear instance name such as:
- `workshop-mlip-yourname`
- `cloud-research-yourname`

A clear name helps you find your instance later, especially if many students are launching machines at the same time.

## Step 3 — Choose the Amazon Machine Image (AMI)

### Recommended option
Choose **Ubuntu Server 22.04 LTS** unless you have already tested another image for the workshop.

### Why Ubuntu
- It is familiar to many researchers.
- Most installation commands used later in the workshop assume an Ubuntu-style package manager.
- It works well for Python-based scientific workflows.

### Avoid for this workshop
- unusual marketplace images,
- images with extra commercial software you do not need,
- operating systems you are not comfortable administering.

## Step 4 — Choose an instance type

### Beginner-friendly recommendation
For a lightweight workshop demo, choose a modest CPU instance first.

Examples of reasonable starting points:
- `t3.medium` for a small general-purpose workshop run,
- a larger CPU instance if you expect a heavier setup,
- a GPU instance only if you have tested the workflow and budget for it.

### How to decide
Ask yourself:
- Is this only for learning the workflow?
- Do I need a GPU today?
- How much cost is acceptable for the workshop?

### Teaching note
For first-time users, it is often better to begin with a simple CPU instance and explain that bigger or GPU-backed instances can be chosen later.

## Step 5 — Create or choose a key pair

1. In the **Key pair** section, either select an existing key pair or click **Create new key pair**.
2. Use a clear name such as `workshop-key-yourname`.
3. Download the key file and save it in a safe place on your laptop.

### Important
- Do not share the private key.
- Do not upload it to GitHub.
- Keep track of the file path because you will need it for SSH.

### Suggested locations
- macOS / Linux: `~/.ssh/workshop-key-yourname.pem`
- Windows: a clearly named folder you can find easily later.

## Step 6 — Configure network settings

In the **Network settings** section, decide how the instance should connect to the network.

### Recommended beginner path
Use an existing default VPC and a default subnet unless you specifically want to teach custom networking.

### What to pay attention to
- **VPC**: your isolated AWS network.
- **Subnet**: where the instance will live inside that VPC.
- **Auto-assign public IP**: this should be enabled for the workshop so you can connect from your laptop.

### Why this matters
For the workshop, the simplest working model is:
- one instance,
- one public IP,
- one security group allowing SSH from your own IP.

## Step 7 — Security group setup

This is one of the most important setup steps.

### Recommended settings
Create a security group that:
- allows **SSH**,
- uses port **22**,
- limits the source to **My IP**.

### What not to do
Avoid opening SSH to the entire internet unless you have a very good reason.

### Suggested security group name
- `workshop-ssh-yourname`

### Why this matters
This rule controls who is allowed to connect to your instance from outside AWS.
For the workshop, the safest simple option is to allow only your own current IP address.

## Step 8 — Configure storage

Look at the **Configure storage** section carefully.

### Main decision
Do you want to keep the root volume after the workshop?

#### Option A — Temporary environment
- Leave the default delete-on-termination behavior.
- Best if you want a clean, disposable setup.

#### Option B — Keep the volume for later reuse
- Change **Delete on termination** to **No** for the root volume.
- Best if you want to preserve installed software, models, trajectories, and outputs after the instance is terminated.

### When keeping the volume makes sense
- You want to continue after the workshop.
- You do not want to repeat package installation.
- You want to preserve the environment as a reusable research setup.

### Warning
Preserved storage can still generate costs even after the compute instance is gone.

## Step 9 — Launch the instance

1. Review all settings.
2. Click **Launch instance**.
3. Wait until AWS reports that the instance launch was successful.
4. Open the instance details page.

### What to check after launch
- instance state changes from **Pending** to **Running**,
- a public IPv4 address appears,
- the correct security group is attached,
- the instance type matches what you intended.

## Step 11 — Sanity checklist before leaving the console

Make sure all of the following are true:

- The instance is in the correct region.
- The instance state is **Running**.
